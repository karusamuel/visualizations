# Introduction to Data Visualization 
## Why Visualization Matters

Data visualization transforms raw numbers into meaningful insights. While statistics such as mean and median are important, visuals allow us to detect patterns much faster than reading tables or spreadsheets.

A well-designed chart can:

- Reveal trends at a glance  
- Communicate insights clearly  
- Support better decision-making  

Humans process visual information far faster than text, making visualization a powerful analytical tool.


## The Power of Visualization

Visualization is not just about presentation — it supports analysis itself.

### 1. Pattern Recognition  
Quickly identify trends, clusters, and anomalies.

### 2. Communication  
Explain findings clearly to non-technical audiences.

### 3. Exploration  
Discover hidden relationships within the data.

### 4. Validation  
Visually confirm statistical results and detect outliers or data quality issues.

## When Visualization Goes Wrong

Poor visualizations can mislead rather than inform.

Common mistakes include:

- Truncated Y-axis — exaggerates small differences  
- Wrong chart type — e.g., too many slices in a pie chart  
- Misleading colors — inconsistent or emotionally confusing color use  
- Chart junk — unnecessary decorations that distract from the data  

**Golden Rule:** Every element in a visualization should serve the data. Clarity is more important than creativity.

## The Visualization Process

Effective visualization follows a structured approach:

1. Define the question you want to answer  
2. Understand your audience  
3. Choose the appropriate chart type  
4. Design for clarity (labels, scales, colors)  
5. Test and refine

# Introduction to Matplotlib

## What Is Matplotlib?

Matplotlib is the foundation of Python visualization.
It offers complete control over every element of a plot. While it requires more effort than higher-level libraries, it provides maximum flexibility and customization.


````python
# import 
import matplotlib.pyplot as plt
````

## Matplotlib Architecture

Understanding the hierarchy is essential.

### 1. Figure  
The overall canvas or window that holds everything.

Think: The entire page.

### 2. Axes  
The actual plotting area where data is drawn.  
A Figure can contain multiple Axes (subplots).

Think: The individual charts on the page.

### 3. Axis  
The number lines (x-axis and y-axis) with ticks and labels.

Think: The rulers with numbers.

Important distinction:
- **Axes** = the full plot area  
- **Axis** = the scale (x or y)

## Two Ways to Use Matplotlib

### 1. Pyplot Interface (MATLAB-style)

- Simpler and quicker
- Less control
- Not ideal for complex or professional projects

Example pattern:
```python
import matplotlib.pyplot as plt
plt.plot(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Title")
plt.show()
```

````python
# example
plt.bar(df_sales["Department"],df_sales["Profit"])
plt.show()
````

````python
# example 2
plt.bar(df_sales["Department"],df_sales["Profit"])
plt.xlabel("Depertment")
plt.ylabel("Profit (Ksh)")
plt.title("Profit Per Deperment in (Ksh)")
plt.show()
````



### 2. Object-Oriented (OO) Interface — Recommended

- Explicit and scalable
- More control
- Best practice for professional work

Example pattern:
```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_title("Title")
plt.show()
```

Best practice:  
Always prefer `fig, ax = plt.subplots()`.


````python
# example
fig,ax = plt.subplots()
ax.bar(df_sales["Department"],df_sales["Profit"])
ax.set_xlabel("Depertment")
ax.set_ylabel("Profit (Ksh)")
ax.set_title("Profit Per Deperment in (Ksh)")
plt.show()
````

## Multiple Subplots

To compare related visuals, create multiple Axes within one Figure.

Example structure:
```python
fig, axes = plt.subplots(2, 2)
axes[0, 0].plot(...)
axes[0, 1].scatter(...)
axes[1, 0].bar(...)
axes[1, 1].hist(...)
plt.tight_layout()
plt.show()
```

Important:
- Use `plt.tight_layout()` to prevent overlapping labels.
- Use `axes[row, column]` for 2D grids.


````python
# example

fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Top-left: Line plot (example)
axes[0, 0].plot(df_sales["Cost"])
axes[0, 0].set_title("Cost Trend")

# Top-right: Scatter plot
axes[0, 1].scatter(df_sales["Cost"], df_sales["Profit"])
axes[0, 1].set_title("Cost vs Profit")
axes[0, 1].grid(True, alpha=0.3)

# Bottom-left: Bar chart
axes[1, 0].bar(df_sales["Department"], df_sales["Cost"])
axes[1, 0].set_title("Cost by Department")
axes[1, 0].tick_params(axis='x', rotation=45)
axes[1, 0].grid(True, axis='y', alpha=0.3)

# Bottom-right: Histogram
axes[1, 1].hist(df_sales["Cost"], color="aqua", edgecolor="black")
axes[1, 1].set_title("Distribution of Cost")
axes[1, 1].grid(True, axis='y', alpha=0.3)

# Adjust layout
plt.tight_layout()
plt.show()
````


## Saving Plots

Export plots for reports or publications:

```python
plt.savefig("chart.png", dpi=300, bbox_inches="tight")
```

Key parameters:
- `dpi=300` → high resolution
- `bbox_inches="tight"` → removes extra whitespace
- `transparent=True` → transparent background
- Can save as PNG, PDF, or SVG


````python
# example 
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Top-left: Line plot (example)
axes[0, 0].plot(df_sales["Cost"])
axes[0, 0].set_title("Cost Trend")

# Top-right: Scatter plot
axes[0, 1].scatter(df_sales["Cost"], df_sales["Profit"])
axes[0, 1].set_title("Cost vs Profit")
axes[0, 1].grid(True, alpha=0.3)

# Bottom-left: Bar chart
axes[1, 0].bar(df_sales["Department"], df_sales["Cost"])
axes[1, 0].set_title("Cost by Department")
axes[1, 0].tick_params(axis='x', rotation=45)
axes[1, 0].grid(True, axis='y', alpha=0.3)

# Bottom-right: Histogram
axes[1, 1].hist(df_sales["Cost"], color="aqua", edgecolor="black")
axes[1, 1].set_title("Distribution of Cost")
axes[1, 1].grid(True, axis='y', alpha=0.3)

# Adjust layout
plt.tight_layout()
plt.savefig("multichart.png",dpi=300,bbox_inches="tight")
````

# Introduction to Seaborn 
## What Is Seaborn?

Seaborn is a statistical visualization library built on top of matplotlib.  it is designed to produce attractive and informative statistical graphics with minimal code.

Seaborn does not replace matplotlib — it enhances it. Under the hood, seaborn generates matplotlib figures, which means you can still customize them using matplotlib tools.



## Why Use Seaborn?

### 1. Beautiful by Default  
Produces professional-looking plots without manual styling.

### 2. Native Pandas Integration  
Works directly with DataFrames. You pass `data=df` and reference columns by name.

### 3. Statistical Focus  
Built-in support for:
- Confidence intervals  
- Regression lines  
- Distributions  
- Aggregations  

### 4. Less Code  
Complex visualizations often require only one or two lines.

## Seaborn vs Matplotlib

### Use Seaborn When:
- Working with DataFrames
- Performing statistical visualization
- Exploring data quickly
- Comparing across categories
- Wanting clean visuals with minimal setup

### Use Matplotlib When:
- You need precise customization
- Creating highly specialized plots
- Adding complex annotations
- Formatting for publication requirements

Best practice: Start with seaborn. If you need deeper customization, modify the resulting matplotlib objects.



## Combining Seaborn and Matplotlib

Since seaborn plots are matplotlib figures, you can combine them:

```python
fig, ax = plt.subplots(figsize=(10, 6))

sns.boxplot(data=df, x="department", y="salary", ax=ax)

ax.set_title("Salary by Department")
ax.axhline(y=75000, linestyle="--")
plt.tight_layout()
plt.show()
```

Use seaborn for plotting and matplotlib for finishing touches.




````python
# example

fig, ax = plt.subplots(figsize=(10, 6))

sns.countplot(data=df_sales,x="Department",ax=ax)

ax.set_title("Salary by Department")
ax.axhline(y=len(df_sales)/6, linestyle="--",color="red")
plt.tight_layout()
plt.show()
````

# Customization & Best Practices  

# The CLARITY Principle

Use this acronym for professional visualizations:

### C — Clear Labels  
Every axis, legend, and title must be self-explanatory.  
No abbreviations without context.

### L — Legible Text  
Font size matters.  
- Minimum 10pt for body text  
- 12–14pt for titles  
Avoid tiny unreadable labels.

### A — Appropriate Colors  
Use color purposefully.  
Consider colorblind viewers. Red and green together is problematic for about 8% of men.

### R — Relevant Data Only  
Remove chart junk, unnecessary gridlines, and 3D effects.  
If it does not improve understanding, delete it.

### I — Intuitive Design  
Follow conventions:
- Time flows left to right  
- Larger values appear higher  
- Positive values go up or right  

### T — Truthful Representation  
Start axes at zero for bar charts.  
Do not manipulate scales to exaggerate results.  
Be honest with data.

### Y — Your Audience First  
Design for viewers, not yourself.  
- Executives need summaries  
- Analysts need detail  

Know your audience.


# Essential Customizations

| Area | Key Principle | When to Use | Best Practices | Example Elements |
|------|--------------|------------|---------------|-----------------|
| Titles & Labels | Provide clear context | All visualizations | Title answers “What am I looking at?”; include units; add year/date | Descriptive title, labeled axes, units in parentheses |
| Sequential Colors | Show ordered progression | Low → high data (temperature, revenue, population) | Use gradual light-to-dark palettes | `Blues`, `YlOrRd`, `viridis` |
| Diverging Colors | Show deviation from midpoint | Correlation, profit vs loss, above/below average | Use contrasting colors around neutral center | `RdBu`, `coolwarm`, `PiYG` |
| Categorical Colors | Distinguish separate groups | Unordered categories | Use clearly distinct hues; ensure colorblind safety | `Set2`, `husl`, `tab10` |
| Accessibility | Ensure inclusivity | All charts | Avoid red–green combos; use colorblind-safe palettes; add patterns/shapes if needed | `Set2` palette |
| Annotations | Highlight insights | When key findings need emphasis | Add reference lines, arrows, labels, shaded regions | `ax.axhline()`, `ax.text()`, `ax.annotate()`, `ax.axvspan()` |
| Gridlines | Improve readability | Most charts | Keep subtle; reduce clutter; adjust transparency | Light dashed horizontal lines |
| Whitespace & Cleanup | Improve clarity | All visualizations | Remove top/right spines; apply `tight_layout()` | `ax.spines[...]`, `plt.tight_layout()` |
| Chart-Specific Grid Rule | Match grid to chart type | Bar vs Scatter | Bar: horizontal only; Scatter: subtle both axes; avoid bold grids | `ax.grid(axis='y', alpha=0.3)` |
